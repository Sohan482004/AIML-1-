```python
x_features=list(df.columns) # 'columns' was misspelled as 'colums'
x_features.remove('status')
encoded_df=pd.get_dummies(df[x_features],drop_first=True) # true should be True (Python is case-sensitive)
print(list(encoded_df.columns))
```

```python
x=encoded_df
y=df['status']
```

```python
!pip install scikit-learn # install the scikit-learn library that contains the train_test_split function
from sklearn.model_selection import train_test_split # import the train_test_split function
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion='gini', max_depth=3)
clf.fit(x_train, y_train)
```

```python
pred_y=clf.predict(x_test)
```

```python
from sklearn import metrics
print("confusion matrix is\n",metrics.confusion_matrix(pred_y,y_test))
print("accuracy is",metrics.accuracy_score(pred_y,y_test))
print("auc score is",metrics.roc_auc_score(pred_y,y_test))
```

```python
from sklearn.tree import export_graphviz
import pydotplus as pdot
from IPython.display import Image
export_graphviz(clf,out_file='tree.dot',feature_names=x_train.columns,filled=True)
graph = pdot.graph_from_dot_file('tree.dot')
graph.write_jpg('tree.png')
Image(filename="tree.png")
```
